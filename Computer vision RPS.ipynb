```python
import random
import cv2
import cvzone
from cvzone.HandTrackingModule import HandDetector
import time

cap = cv2.VideoCapture(0)
cap.set(3, 640)
cap.set(4, 480)

detector = HandDetector(maxHands=1)

timer = 0
stateResult = False
startGame = False
scores = [0,0]
result = " "
ult = 4
gauge = "Gauge0"

while True:
    imgBG = cv2.imread("E:/Python project/RPS.images/BG.png")
    success, img = cap.read()

    imgScaled = cv2.resize(img, (0,0),None,0.9, 0.9)
    imgScaled = imgScaled[:,80:480]

    hands, img = detector.findHands(imgScaled)

    if scores[1] == 0:
        gauge = "Gauge0.png"
    elif scores[1] == 1:
        gauge = "Gauge1.png"
    elif scores[1] == 2:
        gauge = "Gauge2.png"
    elif scores[1] == 3:
        gauge = "Gauge3.png"
    elif scores[1] == 4:
        gauge = "Gauge4.png"
    elif scores[1] == 5:
        gauge = "Gauge5.png"

    if startGame:

        if stateResult is False:
            timer = time.time() - initialTime
            cv2.putText(imgBG,str(int(timer)), (970, 435),cv2.FONT_HERSHEY_PLAIN,6,(255,255,255),4)

            if timer>3:
                stateResult = True
                timer = 0
                
                if hands:
                    playerMove = None
                    hand = hands[0]
                    fingers = detector.fingersUp(hand)
                    if scores[1] >= ult:
                        result = "gun is loaded"
                        if fingers == [1, 1, 0, 0, 0]:
                            playerMove = 4
                            ult += 5
                    if fingers == [0, 0, 0, 0, 0]:
                        playerMove = 1
                    if fingers == [1, 1, 1, 1, 1]:
                        playerMove = 2
                    if fingers == [0, 1, 1, 0, 0]:
                        playerMove = 3
                        
                    randomNumber = random.randint(1, 3)
                    imgAI = cv2.imread(f'E:/Python project/RPS.images/{randomNumber}.png', cv2.IMREAD_UNCHANGED)
                    imgBG = cvzone.overlayPNG(imgBG, imgAI,(300, 310))

                    if playerMove == randomNumber:
                        result = "Tie!"
                    elif (playerMove == 1 and randomNumber == 3) or \
                            (playerMove == 2 and randomNumber == 1) or \
                            (playerMove == 3 and randomNumber == 2):
                        scores[1] += 1
                        result = "You Win!"
                    elif playerMove == 4:
                        scores[1] += 3
                        result = "gun auto-wins"

                    else: 
                        scores[0] += 1
                        result = "You Lose!"


    imgBG[261:693,1345:1745] = imgScaled
    
    if stateResult:
        imgBG = cvzone.overlayPNG(imgBG, imgAI, (300, 310))

    cv2.putText(imgBG, str(scores[0]), (450, 215), cv2.FONT_HERSHEY_PLAIN, 4, (255, 255, 255), 6)
    cv2.putText(imgBG, str(scores[1]), (1512, 215), cv2.FONT_HERSHEY_PLAIN, 4, (255, 255, 255), 6)
    cv2.putText(imgBG, ("Press s to start and q to leave"), (500, 800), cv2.FONT_HERSHEY_PLAIN, 4, (255, 255, 255), 6)
    cv2.putText(imgBG, str(result), (850, 500), cv2.FONT_HERSHEY_PLAIN, 4, (255, 255, 255), 6)
    visualgauge = cv2.imread(f'E:/Python project/RPS.images/Gauge0.png', cv2.IMREAD_UNCHANGED)
#    gaugeoverlay = cvzone.overlayPNG(imgBG, visualgauge,(150,100))
#    resize = cv2.resize(visualgauge, (0.5,1))

    #cv2.imshow("Image", img)
    cv2.imshow("BG", imgBG)
    #cv2.imshow("Scaled", imgScaled)

    
    key = cv2.waitKey(1)
    if key == ord('s'):
        startGame = True
        initialTime = time.time()
        stateResult = False
        result = " "
        
    if key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
